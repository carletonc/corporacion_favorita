# Engineer test and train dataframes to make onpromotion 1-hot and bring in extrapolated date information

```python
import pandas as pd
```

```python
test_dates = pd.read_csv('new_data/test_dates.csv')
test_df = pd.read_csv('data/test.csv')
test_df
```

```python
test_promos = pd.get_dummies(test_df['onpromotion'], drop_first=True)
test_promos.rename(columns={True:'onpromotion'}, inplace=True)
test_df.drop('onpromotion', axis=1, inplace=True)
test_df = test_df.merge(test_promos, left_index=True, right_index=True)
test_df
```

```python
test_df = test_dates.merge(test_df, on='date')
test_df
```

```python
test_df.to_csv('new_data/new_test.csv', index=False)
del test_df, test_promos, test_dates
```

```python
train_dates = pd.read_csv('new_data/train_dates.csv')
train_df = pd.read_csv('data/train.csv')
train_df
```

```python
train_promos = pd.get_dummies(train_df['onpromotion'], drop_first=True)
train_promos.rename(columns={True:'onpromotion'}, inplace=True)
train_df.drop('onpromotion', axis=1, inplace=True)
train_df = train_df.merge(train_promos, left_index=True, right_index=True)
train_df
```

```python
train_df = train_dates.merge(train_df, on='date')
train_df
```

```python
train_df.to_csv('new_data/new_train.csv', index=False)
```
