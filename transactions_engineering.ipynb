```python
import pandas as pd
from collections import Counter
```

```python
training_dates = pd.read_csv('modified_data/training_dates.csv')
training_dates
```

```python
transactions_df = pd.read_csv('data/transactions.csv')
transactions_df
```

```python
# sort by store number
# look at distribution--mean/med/min/max
# throw in chart for insights perhaps?
```

```python
transactions_df.sort_values(['store_nbr', 'date'], inplace=True)
transactions_df
```

```python
trans_groups = transactions_df.groupby('store_nbr')
```

```python
trans_groups.describe(include='all')
```

```python
stores_df = training_dates

for name, group in trans_groups:
    suff = '_'+str(name)
    new_name = 'transactions'+suff
    group = group.drop('store_nbr', axis=1)
    group = group.rename(columns={'transactions': new_name})
    stores_df = stores_df.merge(group, on='date', how='left')

stores_df
```
