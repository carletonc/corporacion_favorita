```python
city_df = pd.get_dummies(stores_df['city'], prefix='city_')
state_df = pd.get_dummies(stores_df['state'], prefix='state_')
type_df = pd.get_dummies(stores_df['type'], prefix='type_')
cluster_df = pd.get_dummies(stores_df['cluster'], prefix='cluster_')
```

```python
stores_df = stores_df.merge(city_df, left_index=True, right_index=True)
stores_df = stores_df.drop(['city'], axis=1)
stores_df = stores_df.merge(state_df, left_index=True, right_index=True)
stores_df = stores_df.drop(['state'], axis=1)
stores_df = stores_df.merge(type_df, left_index=True, right_index=True)
stores_df = stores_df.drop(['type'], axis=1)
stores_df = stores_df.merge(cluster_df, left_index=True, right_index=True)
stores_df = stores_df.drop(['cluster'], axis=1)
stores_df
```

```python
stores_df.to_csv('modified_data/new_stores.csv', index=False)
```
