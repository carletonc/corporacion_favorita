```python
'''
Since date ranges seem to be inconsistant across data, 
let's create two data frames with continuous calendar dates to refer to.

all date ranges 2013-01-01 to 2017-08-31
training date ranges 2013-01-01 to 2017-08-15
testing date ranges 2017-08-16 to 2017-08-31
'''

import pandas as pd
```

```python
training_dates_df = pd.DataFrame(pd.date_range('2013-01-01', '2017-08-15'))
training_dates_df.columns = ['date']
training_dates_df
```

```python
# df['DOB1'] = df['DOB'].dt.strftime('%m/%d/%Y')

training_dates_df['month'] = training_dates_df['date'].dt.strftime('%m')
training_dates_df['day'] = training_dates_df['date'].dt.strftime('%d')
training_dates_df['year'] = training_dates_df['date'].dt.strftime('%Y')

training_dates_df
```

```python
training_dates_df.to_csv('modified_data/train_dates.csv', index=False)
```

```python
testing_dates_df = pd.DataFrame(pd.date_range('2017-08-16', '2017-08-31'))
testing_dates_df.columns = ['date']
testing_dates_df
```

```python
testing_dates_df['month'] = testing_dates_df['date'].dt.strftime('%m')
testing_dates_df['day'] = testing_dates_df['date'].dt.strftime('%d')
testing_dates_df['year'] = testing_dates_df['date'].dt.strftime('%Y')

testing_dates_df
```

```python
testing_dates_df.to_csv('modified_data/test_dates.csv', index=False)
```
