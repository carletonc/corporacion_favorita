```python
import pandas as pd
```

# Merge Test Dataframes

```python
# test_dates = pd.read_csv('test_dates.csv')
test_df = pd.read_csv('new_test.csv')
items = pd.read_csv('new_items.csv')                                    # MERGE ON  --item_nbr
test_store_transactions = pd.read_csv('test_store_transactions.csv')    # MERGE ON -- date, month, day, year
test_holidays = pd.read_csv('test_holidays.csv')                        # MERGE ON -- date, month, day, year
test_oil = pd.read_csv('test_oil.csv')                                  # MERGE ON -- date, month, day, year
```

```python
test_df
```

```python
test_df = test_df.merge(items, on='item_nbr')
test_df.sort_values(by=['id'], inplace=True)
test_df.reset_index(drop=True, inplace=True)
test_df = test_df.merge(test_store_transactions, on=['date','month','day','year'])
test_df = test_df.merge(test_holidays, on=['date','month','day','year'])
test_df = test_df.merge(test_oil, on=['date','month','day','year'])
test_df
```

```python
test_df.to_csv('TEST_FINAL.csv', index=False, chunksize=200000)
```

# Merge Without Items DF

```python
import pandas as pd
# test_dates = pd.read_csv('test_dates.csv')
test_df = pd.read_csv('new_test.csv')
test_store_transactions = pd.read_csv('test_store_transactions.csv')    # MERGE ON -- date, month, day, year
test_holidays = pd.read_csv('test_holidays.csv')                        # MERGE ON -- date, month, day, year
test_oil = pd.read_csv('test_oil.csv')                                  # MERGE ON -- date, month, day, year
```

```python
test_df = test_df.merge(test_store_transactions, on=['date','month','day','year'])
test_df = test_df.merge(test_holidays, on=['date','month','day','year'])
test_df = test_df.merge(test_oil, on=['date','month','day','year'])
test_df
```

```python
test_df.drop('oil/100', axis=1, inplace=True)
test_df.drop('date', axis=1, inplace=True)
#test_df.fillna(0, inplace=True)
```

```python
columns = test_df.columns.values

for c in columns:
    new_c = c
    new_c.replace(' ', '_')
    test_df.rename(columns={c:new_c}, inplace=True)
```

```python
test_df.to_csv('TEST_FINAL_NO_ITEMS.csv', index=False, chunksize=200000)
```
